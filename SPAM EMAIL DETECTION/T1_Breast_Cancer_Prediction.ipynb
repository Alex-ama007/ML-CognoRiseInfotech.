```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
```

```python
x=df.drop(['diagnosis'],axis=1).values
x
```

```python
y=df['diagnosis'].values
y
```

```python
ax =sns.countplot(x=y,data=df)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 5),
                textcoords = 'offset points')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#model creation
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay
knn=KNeighborsClassifier(n_neighbors=7)
tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
rf=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
lg=LogisticRegression()
lst=[knn,tree,rf,lg]
```

```python
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  cmd=ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred),display_labels=pd.unique(y_train))
  print('\n')
  print('Model is ',i)
  print('\n')
  print(y_pred)
  print('!'*100)
  print('score is',accuracy_score(y_test,y_pred))
  print('*'*100)
  print('confusion matrix ',confusion_matrix(y_test,y_pred))
  print('#'*100)
  print('classification_rp',classification_report(y_test,y_pred))
  print('CM display is',cmd.plot())
```
